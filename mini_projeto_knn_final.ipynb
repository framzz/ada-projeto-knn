# **Solução**

### Passo 1 - Definindo o valor de K

```python
k = 3
```

### Passo 2 -  Definir os K vizinhos mais próximos do ponto a ser classificado de acordo com a função de distância


```python
def calcular_distancia(pessoa, listaClassificados):

    distancias = [] # criando uma lista vazia para receber as distâncias
    for item in listaClassificados:

        resultado = (sum(tuple((a - b)**2 for a, b in zip(list(pessoa[2]),list(item[2])))))**0.5
        tupla = (resultado,item[1]) # Adiciona a distancia e a classe para essa pessoa
        distancias.append(tupla) # adicionando as distâncias a lista vazia
    return distancias # retornando a lista de distâncias como resultado
```

### Passo 3 - Problema de Classificação: função para calcular a  **moda**



```python
def calcula_moda(distancias):
  frequencia = {}
  #Percorre as classes que estão presente na lista dos k vizinhos mais próximos.
  # Verifica se essa classe já está no dicionário de frequencias.
  for classe in distancias:
    if classe in frequencia:
      frequencia[classe]=+1
    else:
      frequencia[classe] = 1

  moda = max(frequencia, key=frequencia.get) # Retorna a classe que mais aparece dentre os vizinhos mais
  return moda
```

### Passo 4 - Classificar os registros

```python
# função para classificar um novo registro,
# recebendo de parametros os dados já classificados, os novos dados e o k
def classificar_novo_registro(dados_classificados, novos_dados, k):
    for pessoa in novos_dados:
      vizinhos = calcular_distancia(pessoa, dados_classificados)
      vizinhos.sort() # ordenando as distâncias da menor para a maior
      vizinhos = vizinhos[:k] #pega os primeiros k vizinhos

      #chama a função que calcula a moda, ou seja a classe que mais aparece
      moda = calcula_moda(vizinhos)
      print(f'{pessoa[0]}, {moda[1]}, {pessoa[2]}')


# chama a função para calcular em que classe os novos registros entrariam
classificar_novo_registro(data, no_class, k)
```
